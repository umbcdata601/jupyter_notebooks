# machine learning topics

```python
mlt = Digraph(comment="machine learning")
mlt.clear()
mlt.graph_attr['rankdir'] = 'TB'

mlt.edge('machine learning','supervised (user provides \nexamples of "right answers")')
mlt.edge('machine learning','unsupervised (no labeled data)')

mlt.edge('supervised (user provides \nexamples of "right answers")','Regression (predict \nreal valued output)')

mlt.edge('supervised (user provides \nexamples of "right answers")','Classification (predict\n discrete valued output)')

mlt.edge('Regression (predict \nreal valued output)','Linear regression')

mlt.edge('Classification (predict\n discrete valued output)','Logistic regression')

mlt.edge('Classification (predict\n discrete valued output)','k-nearest neighbors')

mlt.edge('Classification (predict\n discrete valued output)','decision tree')
mlt.edge('Classification (predict\n discrete valued output)','neural network')

mlt.edge('neural network','Recurrent neural network')

mlt.edge('neural network','Convolutional Neural Networks')

mlt.edge('unsupervised (no labeled data)','clustering')

mlt.edge('clustering','Hierarchical clustering')

mlt.edge('clustering','Partitional clustering')

mlt.edge('Hierarchical clustering','Divisive')

mlt.edge('Hierarchical clustering','agglomerative')

mlt.edge('Partitional clustering','k-means')
```

# generic programming task decomposition

```python
tsk = Digraph(comment="task decomposition")
tsk.clear()
tsk.graph_attr['rankdir'] = 'TB'

tsk.edge('given a task','decompose\ninto subtasks')
tsk.edge('decompose\ninto subtasks','for each subtask')
tsk.edge('for each subtask','has someone else\ndone this before?',label='ask')

tsk.edge('has someone else\ndone this before?','invest my time in',label='yes')
tsk.edge('has someone else\ndone this before?','decompose\ninto subtasks',label='no')

tsk.edge('has someone else\ndone this before?','implement\nthe task',label='no')

tsk.edge('invest my time in','implement\nthe task')

tsk.edge('invest my time in','look for a library')

tsk.edge('look for a library','found library')

tsk.edge('look for a library','no library found')

tsk.edge('no library found','implement\nthe task')

tsk.edge('found library','use existing library')

tsk.edge('use existing library','implement\nthe task')
```

```python
tsk.format = 'png'
tsk.render('task_decomposition', view=True)
Image("task_decomposition.png")
```
